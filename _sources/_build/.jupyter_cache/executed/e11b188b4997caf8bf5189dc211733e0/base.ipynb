```matlab
R1 = 200*10^3; % 200 kOhm
R2 = 40*10^3;  %  40 kOhm
R3 = 50*10^3;  %  50 kOhm

C1 = 25*10^(-9); % 25 nF
C2 = 10*10^(-9); % 10 nF
```

```matlab
den = R1*((1/R1+ 1/R2 + 1/R3 + s*C1)*(s*R3*C2) + 1/R2)
```

```matlab
format long
denH = sym2poly(den)
```

```matlab
numH = -1;
```

```matlab
a = denH(1)
b = denH(2)
```

```matlab
w = 1:10:10000;
Hw = -1./(a*w.^2 - j.*b.*w + denH(3));
```

```matlab
semilogx(w, abs(Hw))
xlabel('Radian frequency w (rad/s')
ylabel('|Vout/Vin|')
title('Magnitude Vout/Vin vs. Radian Frequency')
grid
```
