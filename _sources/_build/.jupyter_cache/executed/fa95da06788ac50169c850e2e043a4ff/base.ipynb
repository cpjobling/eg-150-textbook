```matlab
syms t tau alpha
assume(alpha > 0)

x(t) = heaviside(t); % unit step function
subplot(211)
fplot(x(t)),title('x(t)'),ylim([0,1.25])
h(t) = exp(-alpha*t)*heaviside(t);
subplot(212)
fplot(subs(h(t),alpha,1)),title('h(t)')
```

```matlab
y(t) = int(x(tau)*h(t - tau),tau,-Inf,Inf)
```

```matlab
ya(t) = subs(y(t),alpha,1)
fplot(ya(t))
```
