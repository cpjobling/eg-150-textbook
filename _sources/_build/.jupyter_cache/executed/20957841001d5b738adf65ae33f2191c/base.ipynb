```matlab
omega = 2*pi;
t = linspace(0,30,1000);
Q = 20;
sigma = omega/sqrt(4*Q^2 - 1);
subplot(211)
plot(t,exp(-sigma*t).*cos(omega*t)),title('Q = 20')
Q = 10;
sigma = omega/sqrt(4*Q^2 - 1);
subplot(212)
plot(t,exp(-sigma*t).*cos(omega*t)),title('Q = 10')
```
